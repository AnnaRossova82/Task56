```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
```

```python
im=cv2.imread('cat.jpg')
plt.axis('off')
plt.imshow(im)
plt.show
```

```python
im_RGB = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(im_RGB)
plt.show()
```

```python
B = im_RGB.copy()
B[:,:,(0,1)] = 0
plt.axis('off')
plt.imshow(B)
plt.show()
```

```python
G = im_RGB.copy()
G[:,:,(0,2)] = 0
plt.axis('off')
plt.imshow(G)
plt.show()
```

```python
R = im_RGB.copy()
R[:,:,(1,2)] = 0
plt.axis('off')
plt.imshow(R)
plt.show()
```

```python
img_HSV = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
plt.axis('off')
plt.imshow(img_HSV)
plt.show()
```

```python
#h, s, v = cv2.split(img_HSV)
```

```python
H = img_HSV.copy()
H[:,:,(1,2)] = 255
H_RGB=cv2.cvtColor(H, cv2.COLOR_HSV2RGB)
plt.show()
plt.axis('off')
plt.imshow(H_RGB)
```

```python
S = img_HSV.copy()
S[:,:,0] = 179
S[:,:,2] = 255
S_RGB=cv2.cvtColor(S, cv2.COLOR_HSV2RGB)
plt.show()
plt.axis('off')
plt.imshow(S_RGB)
```

```python
V = img_HSV.copy()
V[:,:,0] = 179
V[:,:,1] = 0
V_RGB=cv2.cvtColor(V, cv2.COLOR_HSV2RGB)
plt.show()
plt.axis('off')
plt.imshow(V_RGB)
```

```python
im_LAB = cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
```

```python
L = im_LAB.copy()
L[:,:,(1,2)] = 0
L_RGB = cv2.cvtColor(L, cv2.COLOR_LAB2RGB)
L_GRAY = cv2.cvtColor(L_RGB, cv2.COLOR_RGB2GRAY)
plt.axis('off')
plt.imshow(L_GRAY, cmap='gray')
plt.show()
```

```python
A = im_LAB.copy()
A[:,:,0] = 127
A[:,:,2] = 0
A_RGB = cv2.cvtColor(A, cv2.COLOR_LAB2RGB)
A_GRAY = cv2.cvtColor(A_RGB, cv2.COLOR_RGB2GRAY)
plt.axis('off')
plt.imshow(A_GRAY, cmap='gray')
plt.show()
```

```python
from matplotlib.colors import ListedColormap
from matplotlib.colors import LinearSegmentedColormap
```

```python
cmap_a = clr.LinearSegmentedColormap.from_list('custom blue', ['Red','Gray', 'Green'], N=256)
plt.imshow(A_GRAY, cmap = cmap_a)
plt.show()
```
